```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Stil curat, academic
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Citire date
df = pd.read_csv("03_cleaning_aggregation_student3.csv")

# Verificare
print(df.head())
```

```python
plt.figure()
sns.lineplot(
    data=df,
    x="year",
    y="gdp_per_capita",
    hue="countryRegionId",
    marker="o"
)

plt.title("Evoluia PIB-ului per capita (20202023)")
plt.xlabel("An")
plt.ylabel("PIB per capita (USD)")
plt.legend(title="ar")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="energy_use_per_capita",
    y="gdp_per_capita",
    hue="countryRegionId",
    style="year",
    s=100
)

plt.title("Relaia dintre consumul de energie i PIB per capita")
plt.xlabel("Consum energie per capita")
plt.ylabel("PIB per capita")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.lineplot(
    data=df,
    x="year",
    y="eia_co2_emissions_4008",
    hue="countryRegionId",
    marker="o"
)

plt.title("Evoluia emisiilor de CO2")
plt.xlabel("An")
plt.ylabel("Emisii CO2")
plt.legend(title="ar")
plt.tight_layout()
plt.show()
```

```python
df_2023 = df[df["year"] == 2023]

plt.figure()
sns.barplot(
    data=df_2023,
    x="countryRegionId",
    y="renewables_share"
)

plt.title("Ponderea energiilor regenerabile (2023)")
plt.xlabel("ar")
plt.ylabel("Energie regenerabil (%)")
plt.tight_layout()
plt.show()
```

```python
vars_corr = [
    "eia_primary_energy_44",
    "eia_energy_intensity_47",
    "eia_co2_emissions_4008",
    "energy_use_per_capita",
    "gdp_per_capita",
    "renewables_share"
]

corr = df[vars_corr].corr()

plt.figure()
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Matricea de corelaie a indicatorilor energetici i economici")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.boxplot(
    data=df,
    x="countryRegionId",
    y="eia_energy_intensity_47"
)

plt.title("Distribuia intensitii energetice pe ri (20202023)")
plt.xlabel("ar")
plt.ylabel("Intensitate energetic")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.lineplot(
    data=df,
    x="year",
    y="renewables_share",
    hue="countryRegionId",
    marker="o"
)

plt.title("Evoluia ponderii energiilor regenerabile")
plt.xlabel("An")
plt.ylabel("Energie regenerabil (%)")
plt.tight_layout()
plt.show()
```
